```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
dataset = pd.read_csv("driver.csv")
x = dataset['mean_dist_day'].values
y = dataset['mean_over_speed_perc'].values
dataset = np.array(list(zip(x,y))).reshape(len(x),2)
```

```python
plt.plot()
plt.xlim([0,100])
plt.ylim([0,50])
plt.title('Dataset')
plt.scatter(x,y)
plt.show()
```

```python
#Code for KMeans
kmeans = KMeans(n_clusters = 2)
kmeans.fit(dataset)
print("K-Mean Cluster Centers : ", kmeans.cluster_centers_)
print("K-Mean Labels : ", kmeans.labels_)

plt.title("K-Mean")
plt.scatter(dataset[:,0], dataset[:,1], c=kmeans.labels_, cmap = "rainbow") 
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color = "yellow")
plt.show()
```

```python
#Code for EM
gmm = GaussianMixture(n_components = 2)
gmm.fit(dataset)
em_predictions = gmm.predict(dataset)
print("\nEM Predictions : ")
print(em_predictions)
print("\n")
print("Mean : \n",gmm.means_)
print("\n")
print("Covariances : \n", gmm.covariances_)
print("\n\n")
print("Dataset :")
print(dataset)

plt.title("Expectation Maximum")
plt.scatter(dataset[:,0], dataset[:,1], c=em_predictions, s=50)
plt.show()
```
